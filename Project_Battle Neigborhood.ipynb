# This Code will be used to connect to Foursquare API, I censored My ID Here
### You will get client id and client secret from Foursquare API
### version is the date 

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius
            )
            
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Category']
    
    return(nearby_venues)
```

# Run the code to get the nearby venues, And Return the District Name so we know that it runs

```python
venues_in_seoul = getNearbyVenues(df1['Name'], df1['Latitude'], df1['Longitude'])
```

```python
venues_in_seoul.head()
```

# Lets Process Seoul Data and Mapping it with Weigh
## After that we will use Clustering with K-Means

```python
# one hot encoding
seoul_onehot = pd.get_dummies(venues_in_seoul[['Venue Category']], prefix="", prefix_sep="")
pd.set_option('display.max_columns',None)


seoul_onehot.head()
```

```python
seoul_onehot['Neighborhood'] = venues_in_seoul['Neighborhood'] 
fixed_columns = [seoul_onehot.columns[-1]] + list(seoul_onehot.columns[:-1])
seoul_onehot = seoul_onehot[fixed_columns]
seoul_onehot.head()
```

```python
seoul_grouped = seoul_onehot.groupby('Neighborhood').mean().reset_index()
seoul_grouped.head()
```

```python
num_top_venues = 10

for hood in seoul_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = seoul_grouped[seoul_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
seoul_venues_sorted = pd.DataFrame(columns=columns)
seoul_venues_sorted['Neighborhood'] = seoul_grouped['Neighborhood']

for ind in np.arange(seoul_grouped.shape[0]):
    seoul_venues_sorted.iloc[ind, 1:] = return_most_common_venues(seoul_grouped.iloc[ind, :], num_top_venues)

seoul_venues_sorted.head()
```

# Clustering
## Now we will use K-Means and find cluster district in seoul

```python
from sklearn.cluster import KMeans
# set number of clusters
kclusters = 5

seoul_grouped_clustering = seoul_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(seoul_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:5] 
```

```python
seoul_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
seoul_venues_sorted
```

```python
df1.columns = ['Neighborhood','Latitude','Longitude']
seoul_merged = df1

seoul_merged = seoul_merged.join(seoul_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

seoul_merged.head()
```

# Lets Plot it Map so we can view it

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
seoul_clusters = folium.Map(location=[37.532600,127.024612],zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(seoul_merged['Latitude'], seoul_merged['Longitude'], seoul_merged['Neighborhood'], seoul_merged['Cluster Labels']):
    label = folium.Popup('Cluster ' + str(int(cluster) +1) + '\n' + str(poi) , parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=50,
        popup=label,
        color=rainbow[int(cluster-1)],
        fill=True,
        ).add_to(seoul_clusters)
        
seoul_clusters
```
